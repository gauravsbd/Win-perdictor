```python
import pandas as pd 
import numpy as np
import seaborn as sns 
```

```python
matches = pd.read_csv('matches.csv')
deliveries=pd.read_csv('deliveries.csv')
```

```python
matches.head()
```

```python
matches.shape
```

```python
matches = matches.dropna(subset=['winner'])
```

```python
matches.shape
```

```python
deliveries['player_dismissed']=deliveries['player_dismissed'].fillna("0")
deliveries['player_dismissed']=deliveries['player_dismissed'].apply(lambda x:x if x=="0" else "1")
deliveries
```

```python
total_score=deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score
```

```python
total_score_df=total_score[total_score['inning']==1]
```

```python
total_score_df['target']=total_score_df['total_runs']+1
total_score_df
```

```python
match_df=matches.merge(total_score_df[['match_id','total_runs','target']],left_on='id',right_on='match_id')
match_df
```

```python
match_df['team1'].unique()
```

```python
teams=['Sunrisers Hyderabad',
       'Mumbai Indians',
       'Royal Challengers Bangalore',
       'Delhi Capitals',
       'Rajasthan Royals',
       'Kolkata Knight Riders',
       'Kings XI Punjab',
       'Chennai Super Kings']
       
       
```

```python
match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderbad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderbad')

match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df[['match_id','city','winner','total_runs','target']]
```

```python
deliveries_df=match_df.merge(deliveries,on='match_id')
```

```python
deliveries_df=deliveries_df[deliveries_df['inning']==2]
```

```python
deliveries_df
```

```python

# Now, perform the cumulative sum operation
deliveries_df['current_score'] = deliveries_df.groupby('match_id')['total_runs_y'].cumsum()

deliveries_df
```

```python
deliveries_df['runs_left']=deliveries_df['total_runs_x']-deliveries_df['current_score']
```

```python
deliveries_df.columns
```

```python
deliveries_df['ball_left']=126-(deliveries_df['over']*6 +deliveries_df['ball'])
deliveries_df
```

```python
deliveries_df['player_dismissed']=deliveries_df['player_dismissed'].astype('int')
wickets=deliveries_df.groupby('match_id')['player_dismissed'].cumsum().values
deliveries_df['wickets']=10-wickets
deliveries_df.tail()
```

```python
deliveries_df['crr']=(deliveries_df['current_score']*6)/(120-deliveries_df['ball_left'])
```

```python
deliveries_df
```

```python
deliveries_df['rr']=(deliveries_df['runs_left']*6)/(deliveries_df['ball_left'])
```

```python
deliveries_df
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
    
```

```python
deliveries_df['result']=deliveries_df.apply(result,axis=1)
```

```python
deliveries_df
```

```python
deliveries_df.columns
```

```python
final_df=deliveries_df[['batting_team','bowling_team','city','runs_left','ball_left','wickets','total_runs_x','crr','rr','result']]
```

```python
final_df=final_df[final_df['ball_left']!=0]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df=final_df.sample(final_df.shape[0])
```

```python
x=final_df.iloc[:,:-1]

y=final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=17)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.pipeline import Pipeline
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'step2__C': [0.01, 0.1, 1, 10, 100],   
    'step2__penalty': ['l1', 'l2'],        
}
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))])
```

```python
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(x_train, y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print(f"Best Parameters: {best_params}")
print(f"Best Cross-Validation Accuracy: {best_score}")
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
